1. Erörtern Sie die Unterschiede zwischen einem Data Warehouse und einem Data Mart 

- Ein Data Mart ist eine kuratierte und Beschränkte Teilmenge von Daten, die häufig speziell für Benutzer von Analysen und Business Intelligence erstellt wird. Data Marts werden häufig als Repository für relevante Informationen für eine Untergruppe von Mitarbeitern oder einen bestimmten Anwendungsfall erstellt. Somit ist der Data Mart eine echte Teilmenge des Data Warehouses.

- Da ein Data Mart eine Teilmenge eines Data Warehouse ist, können Unternehmen Data Marts verwenden, um denjenigen Benutzern Zugang zu den Daten zu gewähren, die sonst keinen Zugang zu den Daten haben. Data Marts können aufgrund ihres kleineren und spezialisierten Designs auch weniger kostspielig in der Speicherung und schneller in der Analyse sein.

- Größe: Ein Data Mart ist in der Regel kleiner als 100 GB; ein Data Warehouse ist in der Regel größer als 100 GB und oft ein Terabyte oder mehr.

- Umfang: Ein Data Mart ist auf einen einzigen Schwerpunkt für einen Geschäftsbereich beschränkt; ein Data Warehouse ist in der Regel unternehmensweit und erstreckt sich über mehrere Bereiche.
- Quellen: Ein Data Mart enthält Daten aus nur wenigen Quellen, ein Data Warehouse speichert Daten aus mehreren Quellen.

2. Ein weiteres Konzept ist der Data Lake. Worin unterscheidet sich ein Data Lake von einem Data Warehouse?

Je nach Anforderungen benötigt ein typisches Unternehmen sowohl ein Data Warehouse als auch einen Data Lake, weil diese auf unterschiedliche Bedürfnisse und Anwendungsfälle abzielen.

Ein Data Warehouse ist eine Datenbank, die für die Analyse relationaler Daten aus Transaktionssystemen und Geschäftsanwendungen optimiert ist. Die Datenstruktur und das Schema werden im Voraus definiert und für schnelle SQL-Abfragen optimiert, deren Ergebnisse in der Regel für operative Berichte und Analysen genutzt werden. Die Daten werden bereinigt, angereichert und umgewandelt, so dass sie als „Single Source of Truth“ fungieren können, der die Benutzer vertrauen können.

Ein Data Lake ist anders, denn er speichert relationale Daten aus Geschäftsanwendungen und nicht-relationale Daten aus mobilen Apps, IoT-Geräten und sozialen Medien. Die Struktur der Daten oder des Schemas wird bei der Datenerfassung nicht definiert. Das bedeutet, dass Sie alle Ihre Daten speichern können, ohne Sie sorgfältig planen oder wissen zu müssen, auf welche Fragen Sie in Zukunft Antworten benötigen. Anhand verschiedener Arten von Datenanalysen wie SQL-Abfragen, Big-Data-Analysen, Volltextsuche, Echtzeitanalysen und Machine Learning lassen sich Erkenntnisse gewinnen.

## Aufgabe 2

Recherchieren Sie Anwendungsfälle von Big Data Applikationen, gehen Sie hierbei insbesondere auf folgende Felder ein:

- Predictive Maintainance 
Ausfälle von Maschinen wie z.B. Züge sind sehr teuer und können zudem auch Leben gefährden. Big Data und IoT werden eingesetzt um Daten zu generieren und zu analysieren. Hierbei können in einigen Fällen Ausfälle vorhergesagt werden bevor sie auftreten. Wichtige Einflussfaktoren sind hier z.B. Herstellungsjahr, Modellreihe, Sensordaten, Fehlermeldungen.
- Machine Learning 
Machine Learning Algorithmen sind auf Daten angewiesen. Um so mehr hochqualitative Daten generiert werden können um so besser verhalten sich die ML Algorithmen. 
- Fraud and Compliance 
Viele Betrugsfälle weisen von der Norm abweichende Muster auf (Outlier): z.B. Überweisungen auf andere Kontinente, wenn diese zuvor nie vorkamen. Oder wenn Daten aus dem internen Netz nach Außen gelangen. Hier kann die Erfassung Auswertung von Daten helfen solche Betrugsfälle zu verhindern. 
- Operational Efficiency 
Mit Big Data können Warenströme analysiert werden. So können z.B. die Return Rate von Produkten analysiert werden oder zukünftige Nachfrage vorhergesagt werden. Oder es können betriebliche Abläufe verbessert werden.
- User Experience 
Mit Big Data können User Gewohnheiten analysiert werden: Präferenzen für Filme, Website aufrufe und z.B. können damit auch unzufriedene Kunden entdeckt werden und bessere Angebote unterbreitet werden.
